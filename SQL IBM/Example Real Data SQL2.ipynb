```python
# The SCHOOLS table contains a large number of columns. How many columns does this table have?
%sql select count(*) as Rows, (select count(*) from SYSCAT.COLUMNS where TABNAME = 'SCHOOLS') as Columns from Schools;
```

```python
# How many Elementary Schools are in the dataset?
%sql select "Elementary, Middle, or High School", count("Elementary, Middle, or High School")  from schools group by "Elementary, Middle, or High School";
```

```python
%sql select Count("Elementary, Middle, or High School") as qnt_ES from schools where "Elementary, Middle, or High School" = 'ES';
```

```python
# What is the highest Safety Score?
%sql select max(SAFETY_SCORE) from schools;
```

```python
# Which schools have highest Safety Score?
%sql select NAME_OF_SCHOOL, SAFETY_SCORE from schools where SAFETY_SCORE = (select max(SAFETY_SCORE) from schools); 
```

```python
# Average of Score in Elementary, Middle, or High School
%sql select "Elementary, Middle, or High School", avg(SAFETY_SCORE) from schools group by "Elementary, Middle, or High School";
```

```python
##### What are the top 10 schools with the highest "Average Student Attendance"?
%sql select name_of_school, AVERAGE_STUDENT_ATTENDANCE from schools order by AVERAGE_STUDENT_ATTENDANCE desc nulls last limit 10;
%sql select name_of_school, AVERAGE_STUDENT_ATTENDANCE from schools WHERE AVERAGE_STUDENT_ATTENDANCE IS NOT NULL order by AVERAGE_STUDENT_ATTENDANCE desc limit 10; 
```

```python
##### Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance
%sql select name_of_school, average_student_attendance from schools order by average_student_attendance asc limit 5; 
```

```python
# Now remove the '%' sign from the above result set for Average Student Attendance column
%sql select name_of_school, Replace (average_student_attendance, '%', '') from schools limit 5;
```

```python
# Which Schools have Average Student Attendance lower than 70%?
%sql select name_of_school, average_student_attendance from schools where average_student_attendance < '70';
%sql SELECT Name_of_School, Average_Student_Attendance from SCHOOLS where CAST ( REPLACE(Average_Student_Attendance, '%', '') AS DOUBLE ) < 70;
```

```python
# Get the total College Enrollment for each Community Area
%sql select COMMUNITY_AREA_NUMBER, sum(COLLEGE_ENROLLMENT) as TOTAL from schools group by COMMUNITY_AREA_NUMBER;
```

```python
# Get the 5 Community Areas with the least total College Enrollment  sorted in ascending order 
%sql select Community_area_name, sum(college_enrollment) as total from schools group by Community_area_name order by total asc limit 5;
```
